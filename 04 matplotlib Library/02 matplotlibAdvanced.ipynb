```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('Databases/iris.csv')
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace({'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2})
```

```python
# cmap and alpha
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

## Plot Size

```python
plt.figure(figsize=(5,4))

plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='jet',alpha=0.5)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.colorbar()
```

## Annotations

```python
batters = pd.read_csv('Databases/batter.csv')
batters.head(5)
```

```python
sample_df = batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

## Horizontal and Vertical lines

```python
plt.figure(figsize=(18,10))

plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])

plt.axhline(130,color='red')
plt.axhline(142,color='green')
plt.axvline(30,color='blue')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

## Subplots

```python
# Different ways to plot graphs
```

```python
fig,ax = plt.subplots(figsize=(5,4))

ax.scatter(batters['avg'],batters['strike_rate'])
ax.set_title('Something')
ax.set_xlabel('Avg')
ax.set_ylabel('SR')
```

```python
# multiple graphs
fig,ax = plt.subplots(nrows=2,ncols=1,sharex=True)

ax[0].scatter(batters['avg'],batters['strike_rate'])
ax[0].set_title('Avg VS SR')
ax[1].scatter(batters['avg'],batters['runs'])
ax[1].set_title('Avg VS Runs')
```

## 3D Line Plot

```python
x = [0,1,5]
y = [0,10,13]
z = [0,13,20]

fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(x,y,z,s=[100,100,100])
ax.plot3D(x,y,z,color='red')
```

## 3D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy = np.meshgrid(x,y)
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Contour Plot
3D to 2D

```python
z = xx**2 + yy**2

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contour(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

## Heatmap

```python
delivery = pd.read_csv('Databases/IPL_Ball_by_Ball_2008_2022.csv')
delivery.head()
```

```python
temp_df = delivery[(delivery['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run'] == 6)]
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
grid
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.colorbar()
```
