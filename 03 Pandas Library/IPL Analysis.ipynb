## Filtering data based on condition

```python
mask = data['city']=='Hyderabad'
```

```python
mask
```

```python
data[mask]
```

```python
data[mask].shape
```

```python
def get_city_match_count(city):
    mask = data['city'] == city
    return data[mask].shape[0]
```

```python
get_city_match_count('Bangalore')
```

```python
get_city_match_count("Rajkot")
```

```python
mask1 = data['city'] == 'Hyderabad'
mask2 = data['date'] > '2010-01-01'

data[mask1 & mask2]
```
