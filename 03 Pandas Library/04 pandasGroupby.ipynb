## Agg method

```python
# passing dict
genres.agg(
    {
        'Runtime':'mean',
        'IMDB_Rating':'mean',
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'mean'
    }
)
```

```python
# passing list
genres.agg(['min','max','sum'])
```

```python
# adding both the syntax
genres.agg(
    {
        'Runtime':['sum','mean'],
        'IMDB_Rating':['mean','sum'],
        'No_of_Votes':'sum',
        'Gross':'sum',
        'Metascore':'mean'
    }
)
```

```python
# apply() -> split / apply / combine
genres.apply(min)
```

```python
# find number of movies sarting with A in each group
def foo(group):
    return (group['Series_Title'].str.startswith('A')).sum()

genres.apply(foo)
```

```python
# find ranking of each movie in the group accordingly to IMDB score
def rank_movies(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group

genres.apply(rank_movies)
```

```python
# find normalized IMDB rating group wise
def normal(group):
    group['norm_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min()) / (group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group

genres.apply(normal)
```

```python
# group on multiple cols
duo = movies.groupby(['Director','Star1'])

# size()
duo.size()

# get_group()
duo.get_group(('Aamir Khan','Amole Gupte'))
```

```python
# find most earning actor -> director combo
duo['Gross'].sum().sort_values(ascending=False)
```

```python
# find best actor and genre combo (in terms of metascore)
movies.groupby(['Star1','Genre'])['Metascore'].mean().reset_index().sort_values('Metascore',ascending=False)
```

```python
# agg on multiple broupby
duo.agg(['min','max'])
```

## Excercise

```python
ipl = pd.read_csv('Databases/deliveries.csv')

ipl.head()
```

```python
ipl.shape
```

```python
# find top 10 batsman in terms of runs
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
# find the batsman with max no of sixes
sixers = ipl[ipl['batsman_runs'] == 6]
sixers.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
# find batsman with most number of 4's and 6's in last 5 over
condition1 = ipl[ipl['over'] > 15]
condition2 = condition1[(condition1['batsman_runs'] == 4) | (condition1['batsman_runs'] == 6)]
condition2.groupby('batsman')['batsman'].value_counts().sort_values(ascending=False).head(1).index[0]
```

```python
# find V Kohli's record agains all teams
condition = ipl[ipl['batsman'] == 'V Kohli']
condition.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
# create a function that can return  the highest score of any batsman
def batsman(batsman):
    return ipl[ipl['batsman'] == batsman].groupby('match_id')['batsman_runs'].sum().sort_values(ascending=False).head(1).values[0]

batsman('V Kohli')
```
