### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import scipy
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
# scatter plot -> axes level function
sns.scatterplot(data = tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
# relplot -> fig level function -> square figure
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
# line plot
gap = px.data.gapminder()
gap
```

```python
# India's lifeExp line plot
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
# axes level function
sns.lineplot(data=temp_df,x='year',y='lifeExp')
```

```python
# using relplot
sns.relplot(data=temp_df, x='year', y='lifeExp', kind='line')
```

```python
# hue -> style
temp_df  = gap[gap['country'].isin(['India','Pakistan','China'])]
temp_df
```

```python
sns.relplot(data=temp_df,x='year', y='lifeExp', kind='line', hue='country')
```

```python
# facet plot
# this will nt work with scatterplot or lineplot
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', col='sex',row='smoker',hue='sex')
```

```python
# col wrap
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year', col_wrap=3)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# figure level -> displot
# axes level -> histplot -> kdeplot -> rugplot
```

```python
# plotting univariate histogram
sns.histplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill', kind='hist')
```

```python
# bins parameter
sns.displot(data=tips,x='total_bill', kind='hist',bins=10)
```

```python
# hue parameter
sns.displot(data=tips,x='tip', kind='hist',hue='sex')
```

```python
# element -> step
sns.displot(data=tips,x='tip', kind='hist',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
# histogram of age
sns.displot(data=titanic,x='age', kind='hist',element='step',hue='sex')
```

```python
# facteting using col and row -> will not work in histplot
sns.displot(data=titanic,x='age', kind='hist',col='sex',element='step')
```

```python
# kdeplot -> continous distribution
sns.kdeplot(data=tips,x='total_bill')
```

```python
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
# Rugplot -> variation by drawing ticks along x and y axis
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

### 2. Matrix Plot

- Heatmap
- Clustermap

```python
# heatmap

# Plot rectangular data as a color-encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# axes level function
plt.figure(figsize=(15,10))
sns.heatmap(temp_df)
```

```python
# annot -> linewidth -> cmmap
df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(10,10))
sns.heatmap(df,annot=True,linewidths=0.5,cmap='viridis')
```
